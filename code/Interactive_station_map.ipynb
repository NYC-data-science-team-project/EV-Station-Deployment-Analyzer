```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
!{sys.executable} -m pip install folium
import folium
```

```python
#Load Data
ev_stations = pd.read_csv(r'C:\Users\kanem\Documents\Capstone project\data\alt_fuel_stations.csv')
```

```python
#CLean df
ev_stations_co = ev_stations[ev_stations['State'] == 'CO']
ev_stations_cleaned = ev_stations_co[['Station Name', 'Street Address', 'City', 'State', 'ZIP', 'EV Pricing','Groups With Access Code','Latitude', 'Longitude']]
```

```python
ev_stations_cleaned
```

```python
# Create a map on Colorado
colorado_map = folium.Map(location=[39.5501, -105.7821], zoom_start=7)
# Add EV stations to the map
for index, row in ev_stations_cleaned.iterrows():
    location = [row['Latitude'], row['Longitude']]
    popup_text = f"{row['Station Name']} ({row['Street Address']}, {row['City']}, {row['State']} {row['ZIP']})"
    marker = folium.Marker(location=location, popup=popup_text, icon=folium.Icon(color='green', icon='bolt', prefix='fa'))
    marker.add_to(colorado_map)
# Show the map
colorado_map
```
